```python
import sqlite3
import pandas as pd
import numpy as np
import os

#s.getcwd()

file = 'clause 19_2 220V sides'
file = file+".csv"
dataframe1 = pd.read_csv(file)[23:] # ignores invalid rows

# reset the name of the columns
headers = dataframe1.iloc[0]
header = []
for index in range(len(headers)):
    header.append(headers[index])
dataframe1.columns = header

# remove row with column headers and reset index
dataframe1 = dataframe1[1:]
dataframe1 = dataframe1.reset_index(drop=True)

# initialize a sql database
sqldatabase = sqlite3.connect("sqldatabase.db")

# convert panda dataframe to sql database
dataframe1.to_sql("tabel1", sqldatabase, if_exists="replace")

# execute test query
query = '''SELECT * FROM tabel1 LIMIT 100;'''
pd.read_sql_query(query,sqldatabase)
```
